```python
import os
import pandas as pd

# Load all CSV files from the specified folder path
def load_all_temperature_data(folder_path):
    all_data = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            df['YEAR'] = filename[-9:-4]  # Extract year from filename
            all_data.append(df)

    combined_df = pd.concat(all_data, ignore_index=True)
    return combined_df

# Calculate seasonal averages
def calculate_seasonal_averages(df):
    seasons = {
        'Summer': ['December', 'January', 'February'],
        'Autumn': ['March', 'April', 'May'],
        'Winter': ['June', 'July', 'August'],
        'Spring': ['September', 'October', 'November']
    }

    seasonal_averages = {}
    for season, months in seasons.items():
        season_data = df[months]
        seasonal_avg = season_data.mean(axis=1)
        seasonal_averages[season] = seasonal_avg.mean()

    return seasonal_averages

# Main function
def main():
    folder_path = '/content/'  # Files are uploaded to /content/ in Google Colab

    # Load all data
    df = load_all_temperature_data(folder_path)

    # Calculate seasonal averages
    seasonal_averages = calculate_seasonal_averages(df)

    # Show the seasonal averages
    print("Seasonal Averages:")
    for season, avg_temp in seasonal_averages.items():
        print(f"{season}: {avg_temp}")

# Run the main function
main()
```

```python
import os
import pandas as pd

# Load all CSV files from the specified folder path
def load_all_temperature_data(folder_path):
    all_data = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            df['YEAR'] = filename[-9:-4]  # Extract year from filename
            all_data.append(df)

    combined_df = pd.concat(all_data, ignore_index=True)
    return combined_df

# Find warmest and coolest stations
def find_warmest_and_coolest_stations(df):
    # Ensure the temperature columns are numeric and handle errors
    df.iloc[:, 4:] = df.iloc[:, 4:].apply(pd.to_numeric, errors='coerce')  # Convert to numeric, coercing errors to NaN

    # Calculate the mean temperature for each station
    df['Average_Temperature'] = df.iloc[:, 4:].mean(axis=1)

    # Find the warmest and coolest stations
    warmest_station = df.loc[df['Average_Temperature'].idxmax()]
    coolest_station = df.loc[df['Average_Temperature'].idxmin()]

    return warmest_station['STATION_NAME'], warmest_station['Average_Temperature'], coolest_station['STATION_NAME'], coolest_station['Average_Temperature']

# Main function
def main():
    folder_path = '/content/'  # Files are uploaded to /content/ in Google Colab

    # Load all data
    df = load_all_temperature_data(folder_path)

    # Find the warmest and coolest stations
    warmest_station, warmest_temp, coolest_station, coolest_temp = find_warmest_and_coolest_stations(df)

    # Show the results
    print(f"Warmest Station: {warmest_station}, Average Temperature: {warmest_temp}°C")
    print(f"Coolest Station: {coolest_station}, Average Temperature: {coolest_temp}°C")

# Run the main function
main()
```
