```python
from google.colab import files

# Function to encrypt a character
def encrypt_char(c, n, m):
    if 'a' <= c <= 'z':
        if 'a' <= c <= 'm':
            shift = n * m
            return chr((ord(c) - ord('a') + shift) % 26 + ord('a'))
        else:
            shift = n + m
            return chr((ord(c) - ord('a') - shift) % 26 + ord('a'))
    elif 'A' <= c <= 'Z':
        if 'A' <= c <= 'M':
            shift = n
            return chr((ord(c) - ord('A') - shift) % 26 + ord('A'))
        else:
            shift = m ** 2
            return chr((ord(c) - ord('A') + shift) % 26 + ord('A'))
    else:
        return c

# Function to decrypt a character
def decrypt_char(c, n, m, original_c):
    if 'a' <= original_c <= 'z':
        if 'a' <= original_c <= 'm':
            shift = n * m
            return chr((ord(c) - ord('a') - shift) % 26 + ord('a'))
        else:
            shift = n + m
            return chr((ord(c) - ord('a') + shift) % 26 + ord('a'))
    elif 'A' <= original_c <= 'Z':
        if 'A' <= original_c <= 'M':
            shift = n
            return chr((ord(c) - ord('A') + shift) % 26 + ord('A'))
        else:
            shift = m ** 2
            return chr((ord(c) - ord('A') - shift) % 26 + ord('A'))
    else:
        return c

# Encrypt the entire text
def encrypt_text(text, n, m):
    return ''.join(encrypt_char(c, n, m) for c in text)

# Decrypt the entire text
def decrypt_text(encrypted_text, original_text, n, m):
    return ''.join(decrypt_char(c, n, m, orig) for c, orig in zip(encrypted_text, original_text))

# Check if the decrypted text matches the original text
def check_correctness(original_text, decrypted_text):
    return original_text == decrypted_text

# Main function
def main():
    # Prompt the user to upload the original text file
    print("Please upload the original text file (raw_text.txt) for encryption and decryption.")
    uploaded = files.upload()

    # Take user inputs for n and m
    n = int(input("Enter value for n: "))
    m = int(input("Enter value for m: "))

    # Read the uploaded raw text file
    with open("raw_text.txt", "r", encoding="utf-8") as file:
        raw_text = file.read()

    # Show the original text preview
    print("\nOriginal Text Preview:")
    print(raw_text[:300])  # Display first 300 characters for preview

    # Encrypt the text
    encrypted = encrypt_text(raw_text, n, m)

    # Show the encrypted text preview
    print("\nEncrypted Text Preview:")
    print(encrypted[:300])  # Display first 300 characters for preview

    # Save the encrypted text to a new file
    with open("encrypted_text.txt", "w", encoding="utf-8") as file:
        file.write(encrypted)
        print("\nEncryption complete. Encrypted text saved to 'encrypted_text.txt'.")

    # Automatically download the encrypted text file
    files.download('encrypted_text.txt')

    # Decrypt the text using the original raw text as a guide
    decrypted = decrypt_text(encrypted, raw_text, n, m)

    # Show the decrypted text preview
    print("\nDecrypted Text Preview:")
    print(decrypted[:300])  # Display first 300 characters for preview

    # Check if the decryption is correct
    if check_correctness(raw_text, decrypted):
        print("\nDecryption verified successfully!")
    else:
        print("\nDecryption failed. The texts do not match.")

    # Save the decrypted text to a new file
    with open("decrypted_text.txt", "w", encoding="utf-8") as file:
        file.write(decrypted)
        print("\nDecrypted text saved to 'decrypted_text.txt'.")

    # Automatically download the decrypted text file
    files.download('decrypted_text.txt')


# Run the main function
if __name__ == "__main__":
    main()
```
